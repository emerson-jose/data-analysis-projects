```python
#IMPOTANCAO DAS BIBLIOTECAS

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARREGAMENTO DO ARQUIVO QUE SERA ANALIZADO
df=pd.read_csv('/content/Clientes.csv')
```

```python
df.head(4)
```

```python
#VISUALIANDO INFORMACOES SOBRE O DF
df.info()
```

```python
#VISUALIZANDO A DIMENSAO DO DF
df.shape
```

```python
#IREI DETERMINAR UMA COLUNA DO DF PARA  SER INDICE NO CASO O (ID)
df.set_index('id')
```

```python
# ANTES DE COMECAR O TRATAMENTO IREI TROCAR  O NOME DAS COLUNAS MAS ANTES VOU  VISUALIZA-LAS
df.columns
```

```python
#RENOMEANDO NOMES DAS COLUNAS

df = df.rename(columns={'id':'cod','created_at':'data_de_criação',
'first_name':'NOME', 'last_name':'SOBRE_NOME',
'email':'GMAIL', 'cell_phone':'NUM_CONTATO','country':'PAIS', 'state':
'ESTADO', 'street':'RUA','number':'NUMERO', 'additionals':'ADICIONAIS'})
```

```python
#VISUALISANDO NOVAMENTE O NOME DAS COLUNAS
df.columns
```

```python
# ANTES DE TRATAR VALORES NULOS VOU DEIXAR NOME DAS COLUNAS EM CAIXA BAIXA
df.columns=df.columns.str.lower()
df.info()
```

```python
#AGORA VISUALIZANDO VALORES NULOS

df.isna().sum()
```

```python
# TRATANDO VALORES NAN OU NULOS
# TEMOS 5 COLUNAS COM VALORES NULOS

df['gmail'] = df['gmail'].fillna('NAO_INFORMADO')

df['estado']= df['estado'].fillna('NAO_INFORMADO')

df['rua'] = df['rua'].fillna('NAO_INFORMADA')

df['numero'] = df['numero'].fillna(0)

df['adicionais'] =df['adicionais'].fillna('SEM_ADICIONAL')
```

```python
#visualizar mais uma vez para ver se houve o tratamento

df.isna().sum()
```

```python
#CONTANDO OCORRENCIAS DISTINTAS NAS COLUNAS
df.nunique()
```

```python
df.head(10)
```

```python
#localizando se possui alguma linha  faltante

df[df.isna().any(axis=1)]
```

```python
# VISUALIZANDO COM GRAFICOS

plt.rcParams['figure.figsize']=(13,8) # tamanho do grafico

tops = df['estado'].value_counts().head(6).index

q2=df[df['estado'].isin(tops)].groupby(['estado','estado']).size().unstack() #filtra e agrupa

q2= q2.loc[q2.sum(axis=1).sort_values(ascending=False).index]# ORDENA PELA SOMA TOTAL DOS ESTADOS (MAIOR PARA MENOR )

q2.plot(kind='bar',stacked = True)
plt.title('TOP 6 ESTADOS COM MAIS  FREQUENCIA  ')
plt.ylabel('QUANTIDADE')
plt.xlabel('ESTADOS')
plt.grid(alpha=0.3,color='GREEN')
plt.xticks(rotation=45)
plt.show()
```

```python
# FREQUENCIA COM QUE CADA ESTADO MAIS APARECEM
df['estado'].value_counts()
```

```python
plt.rcParams['figure.figsize']=(15,8) # tamanho do grafico

q10= df['estado'].value_counts().head(5)
q10 =q10.sort_values(ascending=False)

q10.plot(kind='bar',color='skyblue')
plt.title('TOP 5 COM MAIS FREQUENCIA')
plt.ylabel('QUANTIDADE')
plt.xlabel('ESTADOS')
plt.grid(color='black',alpha=0.2)
plt.xticks(rotation=45)
plt.show()
```
